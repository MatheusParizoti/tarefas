```python
from google.colab import files
uploaded = files.upload()

for arquivo in uploaded.keys():
  df2 = pd.read_csv(f'/content/{arquivo}') #referencia na pasta
```

Leitura e Inspeção dos Dados

```python
df_unido = pd.concat([df1, df2], ignore_index=True)

df_unido.head()
```

```python
# Ajustando para o número correto de colunas
df_unido.columns = [
    'Ranque',
    'Nome',
    'Plataforma',
    'Ano',
    'Gênero',
    'Editora',
    'Vendas_NA',
    'Vendas_EU',
    'Vendas_JP',
    'Vendas_Outros',
    'Vendas_Global'
]

# Exibir as primeiras 5 linhas
df_unido.head()

```

Análise Exploratória

```python
contagem_plataforma = df_unido['Plataforma'].value_counts()

contagem_plataforma.head()
```

```python
media_vendas_por_plataforma = df_unido.groupby('Plataforma') ['Vendas_Global'].mean()

print(media_vendas_por_plataforma)
```

```python
contagem_jogos_genero = df_unido['Gênero'].value_counts()

contagem_jogos_genero.head()
```

Filtragem de Dados

```python
df_unido['Vendas_Global'] = pd.to_numeric(df_unido['Vendas_Global'], errors='coerce')
vg = df_unido['Vendas_Global'].mean()
df_filtro = df_unido.loc[df_unido['Vendas_Global'] > vg]
quantidade = len(df_filtro)

print(quantidade)
```

```python
# Filtra os jogos lançados após 2010
df_ano = df_unido.loc[df_unido['Ano'] > 2010]

# Calcula a média das vendas globais desses jogos
media_vendas = df_ano['Vendas_Global'].mean()

print(media_vendas)
```

Operações Avançadas

```python
df_unido['Vendas_NA_EU'] = df_unido['Vendas_NA'] + df_unido['Vendas_EU']

print(df_unido)
```

```python
media_vendas_ano = df_unido.groupby('Ano')['Vendas_Global'].mean()
print(media_vendas_ano)
```
